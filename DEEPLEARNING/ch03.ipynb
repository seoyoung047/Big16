### 03 미분, 순간 변화율과 기울기

- 직선 AB의 기울기 = $\frac{f(b)-f(a)}{b-a}  = \frac{f(a+\Delta x) - f(a)}{\Delta x}$ = 평균 변화율

- 순간변화율 = $\displaystyle\lim_{\Delta x \to 0}\frac{f(a+\Delta x) - f(a)}{\Delta x} = f'(x)$ 


- 미분 기본 공식
    - $ \{f(g(x))\}' = f'(g(x)) \times g'(x) $

<br>

- 신경망은 결합함수 : $f(g(h( \cdots )))$

### 04 편미분

- 함수 f를 'x에 관해 편미분하라' = $ \frac{\partial f}{\partial x} $


ex) $ f(x,y) = x^2 + yx + a  $ 일때

$ \frac{\partial f}{\partial x} = 2x + y $ 
<br>
<br>
$ \frac{\partial f}{\partial y} = x $
