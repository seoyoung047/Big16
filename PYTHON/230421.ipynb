```python
# 이 DataFrame은 loc와 iloc의 결과 같음

df.loc[0]   # 0번째 행 추출
df.iloc[0]  # 결과 같음
```

```python
df.iloc[:,list(range(0,6,2))].head()
```

```python
# 슬라이싱으로 range대신하기

#    start:end:step
#      [10,11,12] # 13 인덱스
df.iloc[10:13,0:6:2]
```

```python
# [10,11,12,13] # 13포함
df.loc[10:13,['country','year','pop']]
```

```python
df[df['country']=='Korea, Rep.'][['country','year','lifeExp']]
```

```python
# lifeExp = 평균 수명
#          연도별   lifeExp의 평균

df.groupby('year')['lifeExp'].mean()
# DataFrameGroupby|  Series  | Series

# groupby 한 값이 index로 사용됨 = key
# => year가 index이자 key
```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
# => multi index (다중 index) = 'year','continent'
```
